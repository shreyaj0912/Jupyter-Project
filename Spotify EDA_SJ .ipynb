```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_tracks = pd.read_csv('F:/Spotify Datasets/tracks.csv')
df_tracks.head()                       
```

```python
#Check null values in the dataset 
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
# Which artist/musician is the least popular on Spotify, top 10 that are least popular 
sorted_df = df_tracks.sort_values('popularity',ascending =True).head(10)
sorted_df
```

```python
#Descriptive statistics for any datset
df_tracks.describe().transpose()
```

```python
#Len function to know number of columns and rows 
len(df_tracks.columns)
len(df_tracks)
```

```python
# Most popular top 10 popular artists, Analysis : Justin Bieber was the top artist with 100 popularity score , 
# QUERY is the function when you have to enter a condition to a column 
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity',ascending = False).head(10)
most_popular
OR 
# most_popular[:10], to get the top 10 
```

```python
# We use inplace when we want to change in the original dataframe 
df_tracks.set_index('release_date', inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# If you need to know which artist is at any specific position use iloc as syntax  
df_tracks[['artists']].iloc[18]
```

```python
# Change the duration from milliseconds to seconds we use lambda 
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x:round (x/1000))
df_tracks.drop('duration_ms',inplace=True,axis =1)
```

```python
df_tracks.duration.head()
```

```python
# Correlation plot in the form of a Heatmap with all variables. There are some negative correlations, 
#Loudness and energy are highly correlated , more the loudness of song high is the energy of the song 
#https://seaborn.pydata.org/generated/seaborn.heatmap.html
corr_df = df_tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize =(14,6))
heatmap = sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno',linewidths=1,linecolor='Black')
heatmap.set_title('Heatmap correlation between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
# Create sample data , multiplied by 0.004 because i wanted 0.4% of the my entire datsaset 
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
#Regression plot using seaborn library using regplot function, shows that both variables are positively correlated   
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title='Loudness vs Energy correlation')
```

```python
#Regression plot using seaborn library using regplot function, shows that both variables are negatively correlated , 
#if one increases the other decrease , shows downward trend of regression line 
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='b').set(title='Popularity vs acousticness correlation')
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
#pip install --user seaborn== 0.11.0
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind ='hist').set(title='Number of songs per year on Spotify')
```

```python
#Seaborn bar plot : Analysis of the graph below after 2010 the duration of the songs started to decrease 
total_dr = df_tracks.duration 
fig_dims=(18,7)
fig,ax =plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years,y=total_dr,ax=ax, errwidth= False).set(title='Year vs Duration')
plt.xticks(rotation=90)
```

```python
# Seaborn line chart , set_style function 
total_dr = df_tracks.duration 
sns.set_style(style='whitegrid')
fig_dims =(10, 5)
fig, ax =plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x=years,y=total_dr,ax=ax).set(title='Year vs Duration')
plt.xticks(rotation=60)
```

```python
# second dataset
df_genre = pd.read_csv('F:/Spotify Datasets/SpotifyFeatures.csv')
```

```python
df_genre.head()
```

```python
#Bar plot- We can see that Genre of World and Classical have high duration whereas Children's music has the least duration or time 
plt.title('Duration of songs in different genres')
sns.color_palette('rocket',as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel('Duration in milliseconds')
plt.ylabel('Genres')
```

```python
# Top 5 Genres by Popularity 
sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous = df_genre.sort_values('popularity',ascending=False).head(10)
sns.barplot(y='genre',x='popularity',data=famous).set(title ='Top 5 Genres by Popularity')
```
